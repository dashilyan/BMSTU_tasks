# **Лабораторная работа 3. Цепи Маркова с дискретным временем, модель однофазной одноканальной системы обслуживания**

## **Задание 1. Модель деградации сервера с профилактикой и восстановлением**

Крупный вычислительный центр имеет один главный сервер. Администраторы проводят еженедельный мониторинг его состояния. Система может находиться в одном из шести состояний:

- $S_1$ (Отличное): Сервер новый, все компоненты работают с максимальной производительностью, температура в норме, ошибок нет.

- $S_2$ (Хорошее): Незначительное снижение производительности (например, из-за фоновых процессов или начала деградации термопасты), но все задачи решаются без задержек.

- $S_3$ (Удовлетворительное): Появились ошибки в памяти (ECC исправляет), вентиляторы работают на повышенных оборотах, производительность упала на 20%.

- $S_4$ (Критическое): Множественные ошибки ввода-вывода, частые сбои программного обеспечения, производительность упала на 50%. Требуется вмешательство администратора.

- $S_5$ (Профилактика): Сервер намеренно остановлен для планового обслуживания (чистка, замена термопасты, обновление ПО). Это состояние достигается только из $S_3$ по решению администратора.

- $S_6$ (Отказ): Сервер полностью вышел из строя (аппаратный сбой).

Временной шаг — одна неделя. Переходы между состояниями описываются следующими правилами (матрица P составляется на их основе):

**Из $S_1$ (Отличное)**:

- С вероятностью $p_{1,1}$ остается в $S_1$.

- С вероятностью $p_{1,2}$ переходит в $S_2$ (естественная деградация).

- С вероятностью $p_{1,3}$ переходит в $S_3$ (внезапный скачок нагрузки/температуры).

- Переходы в $S_4, S_5, S_6$ невозможны.

**Из $S_2$ (Хорошее)**:

- С вероятностью $p_{2,2}$ остается в $S_2$.

- С вероятностью $p_{2,3}$ переходит в $S_3$ (ухудшение).

- С вероятностью $p_{2,4}$ переходит в $S_4$ (критический сбой).


**Из $S_3$ (Удовлетворительное)**:

- С вероятностью $p_{3,3}$ остается в $S_3$ (балансирует на грани).

- С вероятностью $p_{3,4}$ переходит в $S_4$ (ухудшение до критического).

- С вероятностью $p_{3,5}$ переходит в $S_5$ (администратор замечает проблему и ставит на профилактику).

- С вероятностью $p_{3,6}$ переходит сразу в $S_6$ (внезапный отказ, например, перегрев).

- Переходы в $S_1, S_2$ невозможны.

**Из $S_4$ (Критическое)**:

- С вероятностью $p_{4,4}$ остается в $S_4$ (висит на волоске).

- С вероятностью $p_{4,6}$ переходит в $S_6$ (неизбежный отказ).

- С вероятностью $p_{4,5}$ переходит в $S_5$ (администратор успевает спасти ситуацию и отправить на профилактику).

- Переходы в лучшие состояния ($S_1-S_3$) невозможны.

**Из $S_5$ (Профилактика)**:

Профилактика длится ровно 1 неделю. После нее сервер возвращается как новый.

- С вероятностью $p_{5,1}$ переходит в $S_1$ (успешная профилактика).

- С вероятностью $p_{5,2}$ переходит в $S_2$ (профилактика помогла частично, мелкие проблемы остались).

- Переходы в $S_3, S_4, S_6$ невозможны.

- Остаться в $S_5$ на следующий цикл нельзя (профилактика всегда заканчивается за шаг).

**Из $S_6$ (Отказ)**:

Это поглощающее состояние. Сервер ждет замены. С вероятностью 1 остается в $S_6$.

**Требуется:**

Самостоятельно выбрать вероятности перехода. Построить матрицу переходных вероятностей P размером 6x6, построить размеченный граф состояний. В начальный момент сервер новый $P = (1,0,0,0,0,0)$. Найти вероятности нахождения сервера в каждом из состояний после $k, 2k, 4k$ (например, после 2-х, 4-х и 8-ми) недель работы ($k$ выбрать по своему усмотрению).

Найти вероятность того, что сервер проработает без отказа в течение первых двух месяцев (8 недель). Найти среднее время до первого отказа (MTTF — Mean Time To Failure).  Исключите из цепи состояние  $S_5$ (профилактику), замкнув переходы из $S_3$ и $S_4$ напрямую в $S_1$, или просто убрав $S_5$ и перераспределив вероятности. Сравните MTTF для исходной цепи (с профилактикой) и для цепи, где профилактика отсутствует. Как профилактика влияет на надежность? Предложить аналитическое и численное решение путем моделирования.

**Задание на допбаллы:**

1. Неоднородность: Сделать цепь неоднородной — вероятность перехода из $S_3$ в $S_5$ зависит от того, как давно администратор проводил профилактику (последействие) - моделировать.

2. Ввести в $S_3$ и $S_4$ возможность выбора действия (рисковать и не делать профилактику или делать), попытаться найти алгоритм максимизирующий MTTF - моделировать.



## **Задание 2. Модель одноканальной СМО**

Вычислительный центр фирмы состоит из одного главного сервера коллективного пользования. Число работающих в центре программистов в любой момент времени равно $k>10$. Каждый программист готовит свою программу и через терминал передает ее на сервер для выполнения, куда она сразу попадает. Время подготовки программ имеет экспоненциальное распределение со средним значением $t_{1}$ мин. Программисты могут работать в двух режимах: 1. программист приостанавливает свою работу и ожидает ответа от сервера, 2. программист сразу начинает готовить следующую программу (в реальности так бывает, если компиляция идет в фоне). Таким образом, каждый программист генерирует новый запрос сразу после отправки предыдущего.

Время выполнения программы имеет экспоненциальное распределение со средним значением $t_{2}$ мин. Найти для двух режимов работы программистов (теоретически и экспериментально, с использованием пакета queueing):
 - вероятность того, что программа не будет выполнена сразу же, как только она поступила на терминал;
 - среднее время до получения пользователем результатов реализации;
 - среднее количество программ, ожидающих выполнения на сервере.
 - коэффициент простоя системы.

Параметры $k,t_1,t_2$ выбрать по своему усмотрению.

**Задание на допбаллы**:

Экспериментально разработать алгоритм, минимизирующий длину очереди (например, короткие задачи имеют абсолютный приоритет перед длинными (вытесняющий priority scheduling). Если приходит короткая задача, а сервер выполняет длинную, длинная прерывается и ставится обратно в очередь с сохранением состояния).
